# What this file does

Cleans school enrollment data from 1996-2009 in pdf format to csv.

```python
#tools
import pandas as pd
import functools 
import operator  
```

```python
#function to join tuples
def convertTuple(tup): 
    str = functools.reduce(operator.add, (tup)) 
    return str
```

```python
#variables to iterate
skip_rows = 0
n_rows = 48
skip_rows_2 = 0
i = 0

#increment counter
for i in range(1,39):
    print(i)
    
    #base case
    if i <2:

        skip_rows = 4
        df_final = pd.read_csv('jacqui_data_cleaned.csv', skiprows=skip_rows, nrows=n_rows)
        df_2 = pd.read_csv('jacqui_data_cleaned.csv', skiprows=skip_rows_2, nrows=1)
        df_final['Year']=str(convertTuple(df_2.index[0]))
        var_name = df_2['Connecticut State Department of Education'][0]
        df_final.to_csv(f'{i}.csv')
    
    #iteration
    else:

            
        skip_rows_2 = skip_rows + n_rows + 1 

        skip_rows = skip_rows + n_rows + 5

        #incrementation
        if i == 4:
            n_rows = 35
        elif i ==8:
            n_rows = 47    
        elif i in range(9,13):
             n_rows = 43           
        elif i ==13:
            n_rows = 18
        elif i ==18:
            n_rows = 18
        elif i in range(14,18):
            n_rows = 43       
        elif i in range(19,23):
            n_rows = 40
        elif i == 23:
            n_rows = 30
        elif i in range (24,28):
            n_rows = 40
        elif i == 28:
            n_rows = 34
        elif i == 29:
            n_rows = 41
        elif i in range (30,33):
            n_rows = 41
        elif i == 33:
            n_rows = 28
        elif i in range (34,38):
            n_rows = 43
        elif i == 38:
            n_rows = 16
        else:
            n_rows=48

        
        df_final_1 = pd.read_csv('jacqui_data_cleaned.csv', skiprows=skip_rows, nrows=n_rows)
        df_2 = pd.read_csv('jacqui_data_cleaned.csv', skiprows=skip_rows_2, nrows=1)
        print(df_2)
        try:
            df_final_1['Year']=str(convertTuple(df_2.index[0]))
        except:
            df_final_1['Year']=df_2['Connecticut State Department of Education'][0]


        df_final_1.to_csv(f'{i}.csv')
```
